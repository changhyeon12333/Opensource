```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlt
from sklearn.linear_model import LinearRegression

#csv파일 값만 있기때문에 연도,월 요소값을 넣어 데이터 프레임 생성
Year=[]
Month=[]
Elemente=[]
for i in range(1960,2021):
    for j in range(0,36):
        Year.append(i)
        if j < 12:
            m=j+1
            Month.append(m)
            Elemente.append('평균기온')
        elif j<24:
            m=j-11
            Month.append(m)
            Elemente.append('최고기온')
        elif j<36:
            m=j-23
            Month.append(m)            
            Elemente.append('강수량')            
readed_csv=pd.read_csv('weather.csv',sep=',')
df = pd.DataFrame({'Year':Year[:],
                   'Month':Month[:],
                   'Elemente':Elemente[:],
                   'Value':readed_csv['0']})
df
```

```python
#연도별 평균 강수량과 최고기온 평균기온을 그룹화함
grp_year = df.groupby(['Year','Elemente'])[['Value']].mean()
#표시되는 소수점을 설정해준다
pd.options.display.float_format = '{:.3f}'.format

avg_temp=grp_year.xs('평균기온',level='Elemente') #이중인덱스라 xs를 사용해준다
avg_htemp=grp_year.xs('최고기온',level='Elemente')
avg_pct=grp_year.xs('강수량',level='Elemente')#precipitation강수량


```

```python
#우선 연도별로 서울의 기후가 어떻게 변화되었는지 보자
#구한 연도별 평균의 그래프를그려준다
#온도는 측정값이 섭씨지만 강수량은 mm이다
#그래프를 따로 나누지 않고 왼쪽에는 섭씨를 
#오른쪽은 mm로 y축을 잡아주었다.

plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 4)
plt.rcParams['font.size'] = 12


x=np.arange(1960,2021)
y1 = avg_temp.iloc[:,0]
y2 = avg_htemp.iloc[:,0]
y3 = avg_pct.iloc[:,0]
fig, ax1 = plt.subplots()

ax1.plot(x, y1, color='green',label='Average temp', markersize=1, linewidth=2, alpha=1)
ax1.plot(x, y2, color='red',label='Highest temp', markersize=1, linewidth=2, alpha=1)

ax1.set_ylim(10, 20)
ax1.set_xlabel('Year')
ax1.set_ylabel('℃')
ax1.legend(loc=2)
ax1.tick_params(axis='both', direction='in')



ax2 = ax1.twinx()
ax2.bar(x, y3, color='blue', label='Precipitation', alpha=0.3, width=0.7)
ax2.set_ylim(0, 230)
ax2.set_ylabel('mm')
ax2.tick_params(axis='y', direction='in')
ax2.legend(loc=1)

plt.title("Temperature and precipitation by year in Seoul",size = 15,fontweight='bold')
plt.show()
```

```python
# year_mean.loc[1960,'강수량']
# 연도별로 확인했을시에는 예외적인 상황이 있어 서울의 온도변화가
# 눈에 들어오지 않는다
# 그래서 연대별로 그룹화하여 다시 그래프 그린다

years_1960s=grp_year.iloc[list(range(30))]
years_1970s=grp_year.iloc[list(range(30,60))]
years_1980s=grp_year.iloc[list(range(60,90))]
years_1990s=grp_year.iloc[list(range(90,120))]
years_2000s=grp_year.iloc[list(range(120,150))]
years_2010s=grp_year.iloc[list(range(150,180))]

grp_years_1960s = years_1960s.groupby(['Elemente'])[['Value']].mean()
grp_years_1970s = years_1970s.groupby(['Elemente'])[['Value']].mean()
grp_years_1980s = years_1980s.groupby(['Elemente'])[['Value']].mean()
grp_years_1990s = years_1990s.groupby(['Elemente'])[['Value']].mean()
grp_years_2000s = years_2000s.groupby(['Elemente'])[['Value']].mean()
grp_years_2010s = years_2010s.groupby(['Elemente'])[['Value']].mean()
decade_years = pd.concat([grp_years_1960s,grp_years_1970s,grp_years_1980s
                         ,grp_years_1990s,grp_years_2000s,grp_years_2010s])

Years_list=['1960s','1960s','1960s',
           '1970s','1970s','1970s',
           '1980s','1980s','1980s',
           '1990s','1990s','1990s',
           '2000s','2000s','2000s',
           '2010s','2010s','2010s',]

decade_years.reset_index(drop = False,inplace=True)
# 그래프를 만들 시 쉽게 만들기 위해 0부터 시작하는 인덱스를 넣어줌
decade_years.insert(0,"Years",Years_list[:],True)


decade_years

```

```python
decade_pct=decade_years.iloc[0:18:3]["Value"]
decade_avg_htemp=decade_years.iloc[1:18:3]["Value"]
decade_avg_temp=decade_years.iloc[2:18:3]["Value"]

plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 4)
plt.rcParams['font.size'] = 12


x=list(range(1960,2020,10))
y1 = decade_avg_temp
y2 = decade_avg_htemp
y3 = decade_pct
fig, ax1 = plt.subplots()

ax1.plot(x, y1, color='green',label='Average temp', markersize=1, linewidth=3, alpha=1)
ax1.plot(x, y2, color='red',label='Highest temp', markersize=1, linewidth=3, alpha=1)

ax1.set_ylim(11, 18)
ax1.set_xlabel('Year')
ax1.set_ylabel('℃')
ax1.legend(loc=2)
ax1.tick_params(axis='both', direction='in')



ax2 = ax1.twinx()
ax2.bar(x, y3, color='blue', label='Precipitation', alpha=0.3, width=1.5)
ax2.set_ylim(0, 230)
ax2.set_ylabel('mm')
ax2.tick_params(axis='y', direction='in')
ax2.legend(loc=1)

plt.title("Temperature and precipitation by year in Seoul",size = 15,fontweight='bold')
plt.show()
#그래프를보면 강수량은 특별한 규칙이 없지만 
#평균온도와 최고온도 모두 약 1도 가량 증가한것을 확인할 수 있다.
```

```python
#이제 서울의 월별 기후와 강수량을 분석한다
grp_month = df.groupby(['Month','Elemente'])[['Value']].mean()
grp_month
```

```python
avg_temp_month =grp_month.xs('평균기온',level='Elemente') #이중인덱스 xs사용
avg_htemp_month =grp_month.xs('최고기온',level='Elemente')
avg_pct_month =grp_month.xs('강수량',level='Elemente')


plt.style.use('default')
plt.rcParams['figure.figsize'] = (8, 4)
plt.rcParams['font.size'] = 12


x=list(range(1,13))
y1 = avg_temp_month.iloc[:,0]
y2 = avg_htemp_month.iloc[:,0]
y3 = avg_pct_month.iloc[:,0]
fig, ax3 = plt.subplots()

ax3.plot(x, y1, color='green',label='Average temp', markersize=1, linewidth=3, alpha=1)
ax3.plot(x, y2, color='red',label='Highest temp', markersize=1, linewidth=3, alpha=1)

ax3.set_ylim(-5, 30) #월별 온도차및 강수량 차이가 크기때문에 ylim값도 크게 설정해줌
ax3.set_xlabel('Month')
ax3.set_ylabel('℃')
ax3.legend(loc=2)
plt.xticks(np.arange(0, 14, 1), #월별 구분을 쉽게 하기위해 눈금에 레이블 추가 
           labels=['','Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov','Dec',''])
ax3.tick_params(axis='both', direction='in')



ax4 = ax3.twinx()
ax4.bar(x, y3, color='blue', label='Precipitation', alpha=0.3, width=0.8)
ax4.set_ylim(0, 500)
ax4.set_ylabel('mm')
ax4.tick_params(axis='y', direction='in')
ax4.legend(loc=1)

plt.title("Temperature and precipitation by month in Seoul",size = 15,fontweight='bold')
plt.show()
#그래프를 통하여 월별 강수량과 온도변화를 한눈에 알 수 있다.
```

```python
#머신러닝 
avg_temp_data = avg_temp.reset_index()
avg_temp_data
```

```python
X=avg_temp_data["Year"]
y=avg_temp_data["Value"]
plt.plot(X, y, 'o')
plt.show()
```

```python
line_fitter = LinearRegression()
line_fitter.fit(X.values.reshape(-1,1), y)
```

```python
#2021년도 평균기온을 예측해 보았다.
line_fitter.predict([[2021]])
```

```python
plt.plot(X, y, 'o')
plt.plot(X,line_fitter.predict(X.values.reshape(-1,1)))
plt.show()
```
